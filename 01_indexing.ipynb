# Nodes

```python
from llama_index.core.node_parser import SentenceSplitter

node_parser = SentenceSplitter(chunk_size=100, chunk_overlap=0)
```

```python
nodes = node_parser.get_nodes_from_documents(docs, show_progress=True)
```
